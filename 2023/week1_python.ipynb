```python
import pandas as pd
import numpy as np
```

### 1) Input the data

```python
df = pd.read_csv('data/Week1_input.csv')
```

### 2) Split the Transaction Code to extract the letters at the start of the transaction code. These identify the bank who processes the transaction (help). Rename the new field with the Bank code 'Bank'

```python
df['Bank'] = df.loc[:,'Transaction Code'].str.split('-', n=1, expand=True)[0]

#Anhother approach using regular expression
#df['Bank'] = df['Transaction Code'].str.findall("^[A-Z]+")
```

```python
df['Transaction Code'] = df.loc[:,'Transaction Code'].str.split('-', n=1, expand=True)[1]
```

### 3) Rename the values in the Online or In-person field, Online of the 1 values and In-Person for the 2 values

```python
df.loc[df['Online or In-Person'] == 1, 'Online or In-Person'] = "Online"
df.loc[df['Online or In-Person'] == 2, 'Online or In-Person'] = "In-Person"

# Another way, but using numpy also...

#np.where(df['Online or In-Person'] == 1, 'Online', 'In-Person')
```

### 4) Change the date to be the day of the week

```python
df['Transaction Date'] = df['Transaction Date'].astype("datetime64").dt.day_name()
```

### 5) Different levels of detail are required in the outputs. You will need to sum up the values of the transactions in three ways (help):

    1. Total Values of Transactions by each bank
    2. Total Values by Bank, Day of the Week and Type of Transaction (Online or In-Person)
    3. Total Values by Bank and Customer Code

#### 1. Total Values of Transactions by each bank

```python
df.groupby('Bank').sum().drop('Customer Code', axis=1)

# Or...
# type(df.groupby('Bank')['Value'].sum())
```

#### 2. Total Values by Bank, Day of the Week and Type of Transaction (Online or In-Person)

```python
df.groupby(['Bank','Transaction Date', 'Online or In-Person']).sum().drop('Customer Code', axis=1)
```
